```python
import pandas as pd
import numpy as np

np.random.seed(42)

# Simulate RSSI values from 7 Wi-Fi Access Points
data = np.random.randint(-90, -30, size=(200, 7))

# Simulated indoor locations (rooms)
locations = np.random.choice(['Room_A', 'Room_B', 'Room_C', 'Lab_1'], size=200)

df = pd.DataFrame(
    data,
    columns=[f'AP{i}' for i in range(1, 8)]
)

df['Location'] = locations

df.head()
```

```python
X = df.drop('Location', axis=1)   # Wi-Fi RSSI values
y = df['Location']               # Indoor location
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_encoded = le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42
)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
print("Indoor Localization Accuracy:", accuracy)
```
